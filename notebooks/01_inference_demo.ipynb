# explain-omics — inference demo

This notebook shows how to query the ranked interaction chains using **partial inputs** (microbes and/or pathways).


```python
import os
import pandas as pd
from src.explain_omics_inference import ExplainOmicsInference
```

```python
csv_dir = os.path.join('results','explain_omics','csv')
inf = ExplainOmicsInference(csv_dir)
print('Loaded:', csv_dir)
```

## Microbe-only query
Provide a list of microbial genera and retrieve top pathways / host modules / chains.

```python
res = inf.query(microbes=['Fusobacterium'], pathways=None, top_k=20)
res['top_pathways'].head(10)
```

## Pathway-only query

```python
res = inf.query(microbes=None, pathways=['Glycerophospholipid metabolism'], top_k=20)
res.get('top_host_modules', pd.DataFrame()).head(10)
```

## Combined query (microbe + pathway)
This returns the most specific chains consistent with both.

```python
res = inf.query(microbes=['Fusobacterium'], pathways=['Glycerophospholipid metabolism'], top_k=30)
res['top_chains'].head(10)
```
